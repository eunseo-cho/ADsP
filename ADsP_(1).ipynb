      - 프로젝트 정의 및 계획 수립 (KPI/WBS)

      - 데이터 스토어 설계 (RDBMS/ 하둡, NoSQL)
      - 데이터 수집 및 정합성 점검
      
      
      데이터 분석 (analyzing)
      - 분석용 데이터 준비
      - 텍스트 분석
      - 탐색적 분석 (EDA)
      - 모델링 
      - 모델 평가 및 검증

      시스템 구현 (developing)
      - 설계 및 구현
      
      평가 및 전개 (deploying)
